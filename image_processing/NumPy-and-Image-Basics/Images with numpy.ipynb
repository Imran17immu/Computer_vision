```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
#loading the image
pic = Image.open('cristiano_ronaldo_1.jpg')
```

```python
pic
```

```python
#converting the image into an array
pic_array = np.asarray(pic)
```

```python
pic_array
```

```python
type(pic)
```

```python
type(pic_array)
```

```python
#shape of the array
pic_array.shape
```

```python
#displaying the image from the array
plt.imshow(pic_array)
```

```python
#creating a copy of the array
pic_copy = pic_array.copy()
```

```python
pic_copy.shape
```

```python
#accesing all the R G B contents from the array by array slicing
```

```python
# accessing the Red content
pic_copy[:,:,0]
# accessing the Green content
pic_copy[:,:,1]
# accessing the Blue content
pic_copy[:,:,2]
```

```python
#displaying the red content
plt.imshow(pic_copy[:,:,0])
```

```python
#displaying the red content in gray scale
plt.imshow(pic_copy[:,:,0], cmap="gray")
```

```python
#displaying the Green content
plt.imshow(pic_copy[:,:,1])
```

```python
#displaying the Green content in gray scale
plt.imshow(pic_copy[:,:,1], cmap="gray")
```

```python
#displaying the Blue content
plt.imshow(pic_copy[:,:,2])
```

```python
#displaying the Blue content in gray scale
plt.imshow(pic_copy[:,:,2], cmap="gray")
```

```python
#removing the color green from the image
pic_copy[:,:,1] = 0
```
